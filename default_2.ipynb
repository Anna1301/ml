```python
import pandas as pd 
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score, accuracy_score, recall_score
from sklearn.metrics import precision_score

```

```python
df = pd.read_csv('Data.txt',  encoding = "cp1251", sep = ';')
df.head()
```

```python
df.describe(include = ['object'])
```

```python

stacked = df[[ 'Флаг дефолта по кредиту' ]].stack()
df[['Флаг дефолта по кредиту' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

stacked = df[[ 'Пол клиента' ]].stack()
df[['Пол клиента' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

stacked = df[[ 'Образование клиента' ]].stack()
df[['Образование клиента' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

stacked = df[[ 'Тип товара' ]].stack()
df[['Тип товара' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

stacked = df[[ 'Наличие детей у клиента' ]].stack()
df[['Наличие детей у клиента' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

stacked = df[[ 'Регион выдачи кредита' ]].stack()
df[['Регион выдачи кредита' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

stacked = df[[ 'Семейное положение' ]].stack()
df[['Семейное положение' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

stacked = df[[ 'Оператор связи' ]].stack()
df[['Оператор связи' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

stacked = df[[ 'Является ли клиентом банка' ]].stack()
df[['Является ли клиентом банка' ]] = pd.Series(stacked.factorize()[0], index=stacked.index).unstack()

df
```

```python
X1 = df.drop('Флаг дефолта по кредиту', axis=1)
X = X1.drop( 'Тип товара', axis=1)
y = df['Флаг дефолта по кредиту']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
print(X.shape, y.shape)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)


lrmean_cross_val_score = cross_val_score(lr, X_train, y_train, cv=4).mean()


lr_o_y = lr.predict(X_train)
lr_out_y = lr.predict(X_test)
```

```python
lrtrain_precision_score = precision_score(y_train, lr_o_y, average='micro', zero_division=1)
lrtest_precision_score = precision_score(y_test, lr_out_y, average='micro', zero_division=1)
```

```python
lrtrain_recall_score = recall_score(y_train, lr_o_y, average='micro')
lrtest_recall_score = recall_score(y_test, lr_out_y, average='micro')
```

```python
rf = RandomForestClassifier(n_estimators = 80, max_depth = 8)
rf.fit(X_train, y_train)

rfmean_cross_val_score = cross_val_score(rf, X_train, y_train, cv=4).mean()

rf_o_y = rf.predict(X_train)
rf_out_y = rf.predict(X_test)
```

```python
rftrain_precision_score = precision_score(y_train, rf_o_y, average='micro', zero_division=1)
rftest_precision_score = precision_score(y_test, rf_out_y, average='micro', zero_division=1)
```

```python
rftrain_recall_score = recall_score(y_train, rf_o_y, average='micro')
rftest_recall_score = recall_score(y_test, rf_out_y, average='micro')
```

```python
Gp = pd.DataFrame(
    {
        'LogisticRegression': [lrtrain_precision_score, lrtest_precision_score, lrtrain_recall_score, lrtest_recall_score, lrmean_cross_val_score],
        'RandomForest': [rftrain_precision_score, rftest_precision_score, rftrain_recall_score, rftest_recall_score, rfmean_cross_val_score]
    },
    index=['train_precision_score','test_precision_score', 'train_recall_score', 'test_recall_score', 'mean_cross_val_score'] 
)
Gp
```

У Precision высока цена "ложных тревог" (т.е. отказов в кредите, когда клиент способен его обслуживать), при этом цена пропуска (выдачи кредита неплатежеспособному клиенту) низкая. У Recall наоборот высока цена ошибки, а цена ложной тревоги низкая. Лучше смотреть на обе эти метрики в отдельности, чтобы не потерять клиента, чтобы оценить как финансовые риски от невыплаченных кредитов, так и риск потерять клиента.
